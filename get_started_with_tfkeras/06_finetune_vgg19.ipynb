```python
#导入必要的包
import tensorflow as tf
from tensorflow import keras

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import sklearn
import numpy as np
import pandas as pd
import os,sys,time
```

```python
# include top = false 的时候只保留vgg19的卷积层，而不取全连接层
vgg = keras.applications.vgg19.VGG19(weights='imagenet',include_top=False, input_shape=(224,224,3))
```

```python
# fc layer

top = keras.models.Sequential()
top.add(keras.layers.Flatten(input_shape=vgg.output_shape[1:]))
top.add(keras.layers.Dense(4096,activation='relu'))
top.add(keras.layers.Dropout(0.5))
top.add(keras.layers.Dense(256,activation='relu'))
top.add(keras.layers.Dropout(0.5))
top.add(keras.layers.Dense(2,activation='softmax'))

model = keras.models.Sequential()
model.add(vgg)
model.add(top)

train_datagen = keras.preprocessing.image.ImageDataGenerator(
    rotation_range = 40,width_shift_range = 0.2,height_shift_range = 0.2, rescale = 1/255,shear_range = 20,
    zoom_range = 0.2,horizontal_flip = True,fill_mode = 'nearest',) 
test_datagen = keras.preprocessing.image.ImageDataGenerator(rescale = 1/255,) # 数据归一化 

batch_size = 128

# train_data
train_generator = train_datagen.flow_from_directory(
    './data/cat_vs_dog/train',
    target_size=(224,224),
    batch_size=batch_size)

# test_data
test_generator = test_datagen.flow_from_directory(
    './data/cat_vs_dog/test',
    target_size=(224,224),
    batch_size=batch_size )

# train
model.compile(optimizer=keras.optimizers.SGD(lr=1e-4,momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit_generator(train_generator,steps_per_epoch=len(train_generator),
                    epochs=100,validation_data=test_generator,
                    validation_steps=len(test_generator))

model.save('model.h5')
```
